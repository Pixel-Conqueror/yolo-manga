```python
import os
import cv2
from ultralytics import YOLO

# Dossier contenant les images à traiter
input_folder = 'Demon_Slayer/'
cropped_output_folder = 'cropped_regions/'
os.makedirs(cropped_output_folder, exist_ok=True)  # Créer le dossier principal pour les crops

# Fonction pour découper les zones détectées
def crop_region(image, box, output_path, region_index):
    x1, y1, x2, y2 = map(int, box)  # Coordonnées de la box
    cropped = image[y1:y2, x1:x2]  # Découper la région
    cropped_filename = os.path.join(output_path, f"region_{region_index}.jpg")
    cv2.imwrite(cropped_filename, cropped)  # Sauvegarder la région recadrée
    return cropped_filename

# Parcourir chaque fichier JPEG dans le dossier
for file_name in os.listdir(input_folder):
    if file_name.endswith('.jpg') or file_name.endswith('.jpeg'):  # Vérifier les extensions JPEG
        input_path = os.path.join(input_folder, file_name)  # Chemin complet de l'image
        print(f"Traitement de l'image : {input_path}")

        # Charger l'image originale
        original_image = cv2.imread(input_path)

        # Effectuer la prédiction
        results = model(input_path, conf=0.50)

        # Extraire les bounding boxes détectées
        yolo_boxes = results[0].boxes.xyxy.cpu().numpy()  # Récupérer les coordonnées des boxes
        print(f"Nombre de zones détectées : {len(yolo_boxes)}")

        # Créer un sous-dossier spécifique à cette image
        image_name = os.path.splitext(file_name)[0]  # Nom de l'image sans extension
        image_output_folder = os.path.join(cropped_output_folder, image_name)
        os.makedirs(image_output_folder, exist_ok=True)  # Créer un sous-dossier pour cette image

        # Découper et sauvegarder chaque région détectée
        for i, box in enumerate(yolo_boxes):
            cropped_image_path = crop_region(original_image, box, image_output_folder, i)
            print(f"Région {i+1} sauvegardée : {cropped_image_path}")

print("Traitement terminé. Les régions recadrées sont sauvegardées dans :", cropped_output_folder)
```
